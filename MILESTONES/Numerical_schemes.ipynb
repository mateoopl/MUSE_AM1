```python
import import_ipynb
import Numerical_schemes
import Math


def Euler(U,t1,t2,F):      

    dt = t2 - t1
    return U + dt*F(U)


def RangeKutta4(U,t1,t2,F):

    dt = t2 - t1
    k1 = F(U)
    k2 = F(U + (dt/2)*k1)
    k3 = F(U + (dt/2)*k2)
    k4 = F(U + (dt)*k3)
    return U + (dt/6)*(k1 + 2*k2 + 2*k3 + k4)


def Inverse_Euler(U,t1,t2,F):
    dt = t2-t1
    def G(x):
        return x - U - dt*F(x)
    return Math.Newton(G,U)



def Crank_Nicolson(U,t1,t2,F):
    dt = t2-t1
    a = U + (dt/2)*F(U)
    def G(x):
        return x - a - (dt/2)*F(x)
    return Math.Newton(G,U)

```
